```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def read_spectrum(image_path):
    # Read the image
    image = cv2.imread(image_path)
    
    blue=image[:,:,0]
    green=image[:,:,1]
    red=image[:,:,2]
    
    #plt.imshow(red)
    #plt.imshow(blue)
    # Convert the image to grayscale
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    mid=gray_image.shape[0] //2
    length=gray_image.shape[1]

    # Extract the spectrum
    spectrum = gray_image[mid, :]
    print(gray_image.shape)
    # Map the intensity to wavelength
    wavelengths = np.linspace(400, 650, len(spectrum))
    w=1240/spectrum
    #plt.plot(np.linspace(400, 650, length), red[mid,:], color="r",label="Red")
    #plt.plot(np.linspace(400, 650, length), blue[mid,:],color="green",label="Green")
    #plt.plot(np.linspace(400, 650, length), green[mid,:],color="blue",label="Blue")
    #plt.plot(w,spectrum)
    return spectrum, wavelengths

# Read the spectrum from an image
spectrum, wavelengths = read_spectrum('specnew.jpeg')

# Plot the spectrum
#print(wavelengths,spectrum)
plt.plot(wavelengths, spectrum,color="k",label="Spectrum")
plt.xlabel('Wavelength (nm)')
plt.ylabel('Intensity')
plt.legend()
plt.show()
```
